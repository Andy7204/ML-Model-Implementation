```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Codes/2010_2021_HS2_export.csv'
export_data = pd.read_csv(file_path)
export_data.head(100)

plt.figure(figsize=(10,6))
plt.title("Export from 2010 to 2021")
sns.barplot(x=export_data['year'], y = export_data['value'])
plt.ylabel("Value in crores")
plt.xlabel("Years of export")
```

```python
file_path2 = '/content/drive/MyDrive/Codes/Salary_Data.csv'
salary = pd.read_csv(file_path2)
plt.title("Value of exports per nation per year")
sns.heatmap(data = salary, annot=True)
```

```python
insurance_path ='/content/drive/MyDrive/Codes/insurance.csv'
insurance = pd.read_csv(insurance_path)

sns.scatterplot(x= insurance['bmi'],y = insurance['charges'],hue=insurance['smoker'])
```

```python
sns.regplot(x= insurance['bmi'],y = insurance['charges'])
```

```python
sns.lmplot(x= 'bmi',y = 'charges',hue='smoker',data=insurance)
```

```python
sns.swarmplot(x=insurance['smoker'],
              y=insurance['charges'])
```
